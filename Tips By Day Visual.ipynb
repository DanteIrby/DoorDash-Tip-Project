```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df_tip_by_day = pd.read_csv('TipByDay.csv')

# Column names
week_day_column = 'Week Day'
tip_amount_column = 'Tip Amount'

# Convert 'Tip Amount' to numeric, handling potential errors
df_tip_by_day[tip_amount_column] = pd.to_numeric(df_tip_by_day[tip_amount_column], errors='coerce')

#Order the days in a list
day_order = [ 'Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# Ensure the 'Week Day' column is a categorical type with the specified order
df_tip_by_day[week_day_column] = pd.Categorical(df_tip_by_day[week_day_column], categories=day_order, ordered=True)

# Sort the DataFrame by the ordered 'Week Day' for correct plotting order
df_tip_by_day = df_tip_by_day.sort_values(by=week_day_column)

# Create a bar plot of Tip Amount by Week Day
plt.figure(figsize=(10, 6))
sns.barplot(x=week_day_column, y=tip_amount_column, data=df_tip_by_day, palette='viridis')
plt.title('Average Tip Amount by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip Amount ($)')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.grid(axis='y', alpha=0.75)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
